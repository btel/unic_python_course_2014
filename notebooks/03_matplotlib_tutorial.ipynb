Basic plotting

Anscombe's quartet
------------------

```python
x1 = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
x4 = [8,8,8,8,8,8,8,19,8,8,8]

y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]

x_fit =[4, 20]
y_fit = [5, 13]
```

```python
# set the horizontal space between subplots
# plt.subplots_adjust(hspace=0.5)

ax1 = plt.subplot(221)
plt.plot(x1, y1, 'ko');
plt.plot(x_fit, y_fit, 'r')
plt.xlabel('$x_1$')
plt.ylabel('$y_1$')

plt.subplot(222, sharey=ax1)
plt.plot(x1, y2, 'ko');
plt.plot(x_fit, y_fit, 'r')
plt.xlabel('$x_2$')
plt.ylabel('$y_2$')

plt.subplot(223, sharey=ax1)
plt.plot(x1, y3, 'ko');
plt.plot(x_fit, y_fit, 'r')
plt.xlabel('$x_3$')
plt.ylabel('$y_3$')

plt.subplot(224, sharey=ax1)
plt.plot(x4, y4, 'ko');
plt.plot(x_fit, y_fit, 'r')
plt.xlabel('$x_4$')
plt.ylabel('$y_4$')


plt.tight_layout() #automatically make space for labels

# it is also possible to increase manually the distance between axes
# using plt.subplots_adjust command (see above)

plt.savefig("test.png") # save as raster graphics (PNG)
plt.savefig("test.pdf") # save as vector graphics (PDF)
plt.savefig("test.svg") # export for inkscape
```

Polar plots (`polar`)
-------------------------

```python
n = len(y1)
pi = 3.14
theta = [pi * 2 * i / n for i in range(n+1)]
plt.polar(theta, y1 + [y1[0]], 'b');
```

Error area (`fill_between`)
---------------------------

```python
err_y = 2
# construct min/max bounds
# (it could be more easily done with numpy)
ymin = [y - err_y for y in y1]
ymax = [y + err_y for y in y1]

# x values must be sorted
x = list(range(len(y1)))

plt.plot(x, y1, 'k')
plt.fill_between(x, ymin, ymax, edgecolor='none')
```

Placing text
------------

```python
ax = plt.subplot(111)

plt.plot(x1, y1, 'bo')

#set position in data coordinates
plt.text(7, 10, 'Text (data coordinates)', ha='center', va='bottom')

#set position in normalised axes coordinates
plt.text(0.6, 0.2, 'Text (Axes coordinates)', transform=ax.transAxes)
```

Figure layout
=============

Complex layouts (`GridSpec`)
----------------------------

```python
from matplotlib import gridspec

gs = gridspec.GridSpec(2, 2)

plt.subplot(gs[0, 0])
plt.subplot(gs[0, 1])
plt.subplot(gs[1, :])
```

Nested axes (`GridSpecFromSubplotSpec`)  / plotting with axes methods
--------------------------------------------------------------------

```python
gs = gridspec.GridSpec(2, 2)

gs_left = gridspec.GridSpecFromSubplotSpec(3, 1, gs[:,0])

ax1 = plt.subplot(gs[0,1])
plt.subplot(gs[1,1])

ax2 = plt.subplot(gs_left[0])
plt.subplot(gs_left[1])
ax3 = plt.subplot(gs_left[2])

ax2.plot(x1, y1, 'o')
ax1.imshow(img)
ax3.bar(x1, y2)
```

Creating visual style
===================

Using style arguments
-------------------


```python
plt.plot(x1, y1, 'o', markersize=10, mfc='k', mec='r', mew=2)
plt.xlabel('x', size=16)
plt.ylabel('y', size=16)
plt.title('Anscombe dataset', size=16, weight='bold')
```

Using object methods
--------------------

```python
ax = plt.subplot(111)
line, = ax.plot(x1, y1, 'o')
line.set_markersize(10)
line.set_markerfacecolor('y')
line.set_markeredgewidth(3)
ax.set_axis_bgcolor('ivory')

ticklabels = ax.get_xticklabels()

ticklabels[0].set_size(22)

#setp sets the property on all elements in the list
plt.setp(ticklabels, color='r')
# or equivalently:
#[t.set_color('r') for t in ticklabels]

plt.setp(ticklabels[1:-1], visible=False)
```

Adding grid
-----------

```python
plt.plot(x1, y1, 'ko')
plt.grid(axis='y', which='major', ls='-', lw=0.5)
```

Using transparency
------------------

```python
# add horizontal line
plt.axhline(6, color='g', lw=5)

# add filled areas
plt.fill_between(x, 0, y1, color='r', edgecolor='none', zorder=20, alpha=0.5 )
plt.fill_between(x, 0, y2, edgecolor='none', zorder=10, alpha=0.5)
# zorder - changes the order of plotting (both areas are on top of the line)
# alpha - controls transparency; from 0 (transparent) to 1 (opaque)
```

Configuring spines
------------------

```python
# switch off frame
plt.subplot(111, frameon=False)
plt.plot(x1, y1)
```

```python
ax = plt.subplot(111)

# remove right spine
ax.spines['right'].set_visible(False)

# move left spine outwards
ax.spines['left'].set_position(('outward', 15))

plt.plot(x1, y1)
```
