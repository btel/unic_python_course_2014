Figure layout
=============

Complex layouts (`GridSpec`)
----------------------------

```python
from matplotlib import gridspec

gs = gridspec.GridSpec(2, 2)

gs_topright = gridspec.GridSpecFromSubplotSpec(2, 2, gs[0,0])

for i in range(3):
    plt.subplot(gs[i+1])
for i in range(4):
    plt.subplot(gs_topright[i])
```

Creating visual style
===================

Using style arguments
-------------------


```python
plt.plot(x1, y1, 'o', markersize=10, mfc='k', mec='r', mew=2)
plt.xlabel('x', size=16)
plt.ylabel('y', size=16)
plt.title('Anscombe dataset', size=16, weight='bold')
```

Customising globals
-----------------

```python
import matplotlib
matplotlib.rcParams.update({'xtick.labelsize' : 16,
                            'ytick.labelsize' : 16,
                            'axes.labelsize' : 16,
                            'axes.titlesize' : 18,
                            'lines.markersize' : 10,
                            'lines.linestyle' : 'None',
                            'lines.marker' : 'o',
                            'lines.markerfacecolor' : 'k'})

plt.plot(x1, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Anscombe dataset')
```

Using object-oriented API
-----------------------

```python
ax = plt.subplot(111)
line, = ax.plot(x1, y1, 'o')
line.set_markersize(10)
line.set_mfc('k')

ticklabels = ax.get_xticklabels()

plt.setp(ticklabels, color='r')
plt.setp(ticklabels[1:-1], visible=False)
```
